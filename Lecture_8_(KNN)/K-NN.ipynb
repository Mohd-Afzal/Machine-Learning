# KNN

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
```

```python
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

x_axis = []
k_scores = []

for k in range(1, 50, 2):
    x_axis.append(k)
    clf = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(clf, x_train, y_train, cv = 10, scoring = 'accuracy')
    k_scores.append(scores.mean())
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_axis, k_scores)
plt.show()
```

```python
optimal_k = x_axis[k_scores.index(max(k_scores))]
optimal_k
```

```python
def equcDist(test_data, data):
    dist = []
    
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            euDist = (test_data[j] - data[i][j]) ** 2
        
        dist.append([euDist, i])
     
    sorted(dist)
    return dist
    
```

```python
def fit(x_train, x_test):
    eachTestSortedDist = []
    
    for i in range(x_test.shape[0]):
        eachTestSortedDist.append(equcDist(x_test[i], x_train))
    
    return eachTestSortedDist
```
